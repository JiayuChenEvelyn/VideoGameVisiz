# **Part 2 - Data Preprocessing**

## *Release_Date Processing*

```python
from datetime import datetime

# Convert released_date column to datetime type, handle invalid dates
df['released_date'] = pd.to_datetime(df['released_date'], format="%d %b, %Y", errors='coerce')

# Extract month and year and store in new column
df['released_year'] = df['released_date'].dt.strftime('%Y')
df
```

## *Tags Processing*

```python
from collections import Counter
import ast
# Initialize a Counter object
tag_counter = Counter()

# iterate over each row in the DataFrame
for index, row in df.iterrows():
    tags = ast.literal_eval(row['tags'])
    
    # update Counter 
    tag_counter.update(tags)

```

## *Price Processing*

```python
import re

def convert_price(price):
    if price in ["Free", "Free To Play"]:
        return 0
    else:
        # Use regex to find the numeric part
        matches = re.findall(r'\d+', price)
        if matches:
            return int(matches[0])
        else:
            return 0

df['price'] = df['price'].astype(str).apply(convert_price)


df.head()
```

## *Price*

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Divide the price range
bins = [-float('inf'), 100, 150, 200, 250,300, float('inf')]
labels = ['0-100', '100-150', '150-200', '200-250', '250-300','300+']
df['price_range'] = pd.cut(df['price'], bins=bins, labels=labels, right=False)

# Calculate the number of games in each price range
price_range_counts = df['price_range'].value_counts().sort_index()

# draw histogram
plt.figure(figsize=(10, 6))
sns.barplot(x=price_range_counts.index, y=price_range_counts.values, palette="viridis")
plt.title('Number of Games per Price Range')
plt.xlabel('Price Range')
plt.ylabel('Number of Games')
plt.show()

# draw the pie chart
plt.figure(figsize=(8, 8))
plt.pie(price_range_counts, labels=price_range_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Games in Different Price Ranges')
plt.show()
```

## *Score Range*

```python
# Divide the score into sections
bins = [-float('inf'), 30, 50, 70, 90, 100, float('inf')]
labels = ['0-30', '30-50', '50-70', '70-90', '90-99','100']
df['score_range'] = pd.cut(df['review_score'], bins=bins, labels=labels, right=False)

# Calculate the number of games in each score range
score_range_counts = df['score_range'].value_counts().sort_index()

# Draw a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=score_range_counts.index, y=score_range_counts.values, palette="viridis")
plt.title('Number of Games per Score Range')
plt.xlabel('Score Range')
plt.ylabel('Number of Games')
plt.show()

```

## *Price Range Distribution by Score Range*

```python

# Calculate the number of each price_range in each score_range
grouped = df.groupby(['score_range', 'price_range']).size().unstack()

# Plot stacked bar chart
grouped.plot(kind='bar', stacked=True, figsize=(10, 6))

# Set graph title and axis labels
plt.title('Price Range Distribution by Score Range')
plt.xlabel('Score Range')
plt.ylabel('Count')

plt.show()
```

## *Score Range Proportion by Released Year*

```python
# Calculate the proportion of each score_range in each released_year
grouped = df.groupby(['released_year', 'score_range']).size()
grouped = grouped / grouped.groupby(level=0).sum()

# Unstack score_range level
grouped = grouped.unstack()

# Plot line chart
grouped.plot(kind='line', figsize=(15, 8))

# Set graph title and axis labels
plt.title('Score Range Proportion by Released Year')
plt.xlabel('Released Year')
plt.ylabel('Proportion')

plt.show()
```
